```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
warmup = 2000
max_steps = 30000
power = 3.0

# Learning rate lambda function
def lr_lambda(step: int) -> float:
    min_lr = 0.5
    if step < warmup:
        base = float(step) / float(max(1, warmup))
    else:
        progress = float(step - warmup) / float(max(1, max_steps - warmup))
        base = max(0.0, (1.0 - progress) ** power)

    return min_lr + (1.0 - min_lr) * base

# Compute learning rates for each step
steps = np.arange(0, max_steps + 1)
lrs = [lr_lambda(step) for step in steps]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(steps, lrs)
plt.xlabel('Step')
plt.ylabel('Learning Rate Multiplier')
plt.title('Learning Rate Schedule: Warmup + Polynomial Decay (power=1.0)')
plt.show()
```
